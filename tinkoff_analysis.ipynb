```python
#Let's clean 'events'
a = []
for i in range (len(df)):
    if (df['event'][i] == 'like'):
        a.append('like')
    elif (df['event'][i] == 'dislike'):
        a.append('disl')
    elif (df['event'][i] == 'favorite'):
        a.append('fav')
    else:
        a.append(np.NaN)
```

```python

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Spend 20% of whole amount', 'Spend 80% of whole amount'
sizes = [68.6, 31.4]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, colors=['gold', 'darkgray'], textprops={'fontsize': 14})
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
